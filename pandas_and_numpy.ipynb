```python
q1 = data['Annual Salary'].quantile(0.25)
q3 = data['Annual Salary'].quantile(0.75)

iqr = q3 - q1

ll = q1 - 1.5 * iqr
ul = q3 + 1.5 * iqr

outliers = data[(data['Annual Salary'] < ll) | (data['Annual Salary'] > ul)]
outliers
```
