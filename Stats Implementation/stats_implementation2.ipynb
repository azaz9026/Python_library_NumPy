# Z - test

```python
from numpy.random import  randn
population = randn(100)
```

```python
population
```

```python
sns.distplot(population , kde= True , color='red' )
```

```python
np.mean(population) , np.std(population)
```

```python
null_mean = 0.09
alternate = 0.09
```

```python
from statsmodels.stats.weightstats import ztest
from random import sample

zscore , pvalues = ztest(population , value = null_mean , alternative = 'larger')

zscore 
```

```python
pvalues
```

```python
if pvalues <= 0.05:
    print("Reject the null hypothesis")
else:
    print("fail to Reject the null hypothesis")
```

# T - test

```python
# One Sample t-test----------------------------------------

population = np.random.randint(10 , 50 , 50)
population
```

```python
sample = np.random.choice(population , 20)
print(sample)
```

```python
np.mean(sample)
null_value = 25

import scipy

statistic , pvalue = scipy.stats.ttest_1samp(sample , null_value) 

pvalue
```

```python
# if the P Values is > 0.05 , fail  to reject null hypothesis ---------------------------------

if pvalue >= 0.05:
    print(" fail to Reject the null hypothesis")
else:
    print("Reject the null hypothesis")
```

```python
# two Sample t-test----------------------------------------

azaz_score = np.random.choice(population , 20)

lux_score = np.random.choice(population , 20)
```

```python
np.mean(azaz_score) , np.mean(lux_score)
```

```python
scipy.stats.ttest_ind(azaz_score , lux_score)
```

```python
# if the P Values is > 0.05 , fail  to reject null hypothesis ---------------------------------

if pvalue >= 0.05:
    print(" fail to Reject the null hypothesis")
else:
    print("Reject the null hypothesis")
```

```python
# paired t-test----------------------------------------

azaz_score_1 = np.random.choice(population , 20)
azaz_score_2 = np.random.choice(population , 20)
```

```python
np.mean(azaz_score_1) , np.mean(azaz_score_2)

scipy.stats.ttest_rel(azaz_score_1, azaz_score_2)
```

```python
# if the P Values is > 0.05 , fail  to reject null hypothesis ---------------------------------

if pvalue >= 0.05:
    print(" fail to Reject the null hypothesis")
else:
    print("Reject the null hypothesis")
```

# Chisquare test

```python
 # dataset of tips table-------------------------------------------------

data = sns.load_dataset('tips')
data
```

```python
data_table = pd.crosstab(data['sex'] , data['smoker'])   # using chisquare chi_critical method--------------------------------
data_table
```

```python
observed_values = data_table.values
observed_values
```

```python
import scipy.stats as  stats

stats_test , P_value , dof , expctd_values = stats.chi2_contingency(observed_values)

expctd_values 
```

```python
observed_values
```

```python
dof = 1

from scipy.stats import chi2

for i in zip(observed_values , expctd_values):
    print(i)
```

```python
Chisquare_test =  sum([(o-e)**2/e for o,e in zip( observed_values , expctd_values )])

Chisquare_test
```

```python
Chisquare_stats = Chisquare_test[0] + Chisquare_test[1]

Chisquare_stats
```

```python
alpha  = 0.05 

chi_critical = chi2.ppf(1-alpha , df = dof )

chi_critical
```

```python
if Chisquare_stats >= chi_critical:
    print("Reject the null hypothesis")
else:
    print(" fail to Reject the null hypothesis")
```

```python
# Qus 1 -------------------------------------------------------------

expctd_data = [8,5,6,8,5,7,6]
observed_data = [7,4,7,9,4,8,6]

sum(expctd_data) , sum(observed_data)
```

```python
chisquare_stats , pvalue = stats.chisquare(observed_data , expctd_data)

chisquare_stats , pvalue
```

```python
# find the critical values 

from scipy import stats
alpha = 0.05
dof = len(observed_data) - 1
critical_value = stats.chi2.ppf(1 - alpha, dof)
```

```python
critical_value
```

```python
if chisquare_stats >= critical_value:
    print("Reject the null hypothesis")
else:
    print(" fail to Reject the null hypothesis")
```
