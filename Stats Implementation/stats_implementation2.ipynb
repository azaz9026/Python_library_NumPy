# Central Limit theorem

```python
population = np.random.binomial(10 , 0.5 , 100000)

population
```

```python
len(population)
```

```python
sns.displot(population , kind='kde')
```

```python
sample_size = int(len(population)*.30)

sample_size
```

```python
mean_of_sample = [] 

for i in range(1,1000):
    sample = np.random.choice(population , size= sample_size)
    mean_of_sample.append(np.mean(sample))
```

```python
mean_of_sample
```

```python


sns.distplot(mean_of_sample , kde= True , color='red' , label='Mean of sample')
```

# Z - test

```python
from numpy.random import  randn
population = randn(100)
```

```python
population
```

```python
sns.distplot(population , kde= True , color='red' )
```

```python
np.mean(population) , np.std(population)
```

```python
null_mean = 0.09
alternate = 0.09
```

```python
from statsmodels.stats.weightstats import ztest
from random import sample

zscore , pvalues = ztest(population , value = null_mean , alternative = 'larger')

zscore 
```

```python
pvalues
```

```python
if pvalues <= 0.05:
    print("Reject the null hypothesis")
else:
    print("fail to Reject the null hypothesis")
```
